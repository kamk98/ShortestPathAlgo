```python
def ask():
    global i       #1st Start
    num_edge=input("Enter the number of edges:")
    num_edge1=int(num_edge)
    for i in range(0,num_edge1):          #asking for the edges details num_edge1 times
        fr_edge=input("Enter from vertex:")
        t_edge=input("Enter to vertex:")
        cost1=input("Enter cost:")
        print("\n")
        if(not from_list):
            from_list.append(fr_edge)
            to_list.append(t_edge)
            costint=int(cost1)            
            cost_list.append(costint)
        else:
            for j in range(0,len(from_list)):
                if(fr_edge<from_list[j]):
                    from_list.insert(j, fr_edge)
                    to_list.insert(j,t_edge)
                    costint=int(cost1)
                    cost_list.insert(j,costint)
                    break
                elif(j==len(from_list)-1):
                    from_list.append(fr_edge)
                    to_list.append(t_edge)
                    costint=int(cost1)            
                    cost_list.append(costint)
                    break
    print("After final Optimization")
    print("From list is {}".format(from_list))
    print("To list is {}".format(to_list))
    prin.t("Cost list is {}".format(cost_list))
    source_vertex=input("Enter source vertex:")
    dest_vertex=input("Enter destination vertex:")    #1st End
    j=0                                               #2nd Start
    while(j<len(from_list)):      #searching for source_vertex in from_list
        if(from_list[j]==source_vertex):
            ml1.append(from_list[j])
            ml1.append(to_list[j])
            LKI.append(j)
            if(to_list[j]==dest_vertex):    #In the case that the first occurence of source_vertex corresponds to dest_vertex itself
                ML1.extend(ml1)
                del(ml1[-1])         #performing SIEC of some sort
                keyel=ml1[0]
                i=j+1       #This i is the index from which keyel is to be searched.
                break       #we break from this loop and head straight to function func2 where i's value is one obtained in the line above
            keyel=to_list[j]
            i=0
            break
        elif((j==(len(from_list)-1))&(from_list[j]!=source_vertex)): #worst case that source_vertex is not even found
            print('No path for given vertices')
            break
        else:
            j=j+1
            continue
        #First two nodes are done till here
    func2(source_vertex,dest_vertex,keyel,ml1,ML1,q,i)
    func3(ML1,ml2,source_vertex,dest_vertex,cost_list1)  
```
